Task 2
You are given a customer churn dataset (customer_churn.csv).
1. Preprocess:
   * Encode contract_type
    * Normalize features
2. Train:
    * KNN (k=3, k=5)
    * Decision Tree (default + max_depth=3)
3. Evaluate:
   * Accuracy
    * Confusion Matrix
4. Compare models and justify which performs better

```python
import pandas as pd
import numpy as np

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder, StandardScaler

from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier

from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
```

```python
df = pd.read_csv("/content/churn_data.csv")
print(df.head(25))
```

```python
le = LabelEncoder()
df['contract_type'] = le.fit_transform(df['contract_type'])
```

```python
X = df.drop("churn", axis=1)
y = df["churn"]
```

```python
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
X_train, X_test, y_train, y_test = train_test_split(
    X_scaled, y, test_size=0.35, random_state=42
)
```

KNN = 3

```python
knn3 = KNeighborsClassifier(n_neighbors=3)
knn3.fit(X_train, y_train)
pred_knn3 = knn3.predict(X_test)
```

KNN =5

```python
knn5 = KNeighborsClassifier(n_neighbors=5)
knn5.fit(X_train, y_train)
pred_knn5 = knn5.predict(X_test)
```

Default Decision Tree

```python
dt_default = DecisionTreeClassifier(random_state=42)
dt_default.fit(X_train, y_train)
pred_dt_default = dt_default.predict(X_test)
```

Decision Tree Depth= 3

```python
dt_limited = DecisionTreeClassifier(max_depth=3, random_state=42)
dt_limited.fit(X_train, y_train)
pred_dt_limited = dt_limited.predict(X_test)
```

```python
def evaluate_model(name, y_true, y_pred):
    print(f"\n===== {name} =====")
    print("Accuracy:", accuracy_score(y_true, y_pred))
    print("Confusion Matrix:\n", confusion_matrix(y_true, y_pred))
    print("\nClassification Report:\n", classification_report(y_true, y_pred))
```

```python
evaluate_model("KNN (k=3)", y_test, pred_knn3)
evaluate_model("KNN (k=5)", y_test, pred_knn5)

evaluate_model("Decision Tree (Default)", y_test, pred_dt_default)
evaluate_model("Decision Tree (max_depth=3)", y_test, pred_dt_limited)
```

```python
results = {
    "KNN (k=3)": accuracy_score(y_test, pred_knn3),
    "KNN (k=5)": accuracy_score(y_test, pred_knn5),
    "Decision Tree (Default)": accuracy_score(y_test, pred_dt_default),
    "Decision Tree (max_depth=3)": accuracy_score(y_test, pred_dt_limited)
}

best_model = max(results, key=results.get)

print("\n📊 Model Accuracy Comparison:")
for model, acc in results.items():
    print(f"{model}: {acc:.4f}")

print("\n🏆 Best Performing Model:", best_model)
```
