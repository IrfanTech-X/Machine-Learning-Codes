Generate a synthetic dataset of 20 points (2d) with 2 classes using any rule. split into train 70% and test 30%.

Train :
* KNN(K=3 and K=5)
* Decision tree (default and max_depth =3) .

Compute test accuracy(before and after noise).


Output:

Before Noise:
* KNN(K=3): ..
* KNN(k=5): ..
* Decision tree(default): ..
* Decision tree(depth=3): ..

After Noise:
* KNN(K=3): ..
* KNN(k=5): ..
* Decision tree(default): ..
* Decision tree(depth=3): ..

```python
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score
```

Generating synthetic dataset

```python
np.random.seed(42)

# Class 0: centered around (1,1)
class0 = np.random.randn(10, 2) + [1, 1]

# Class 1: centered around (5,5)
class1 = np.random.randn(10, 2) + [5, 5]

X = np.vstack((class0, class1))
y = np.array([0]*10 + [1]*10)
```

```python
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.3, random_state=42)
```

Training model before(noise)

```python
knn3 = KNeighborsClassifier(n_neighbors=3)
knn5 = KNeighborsClassifier(n_neighbors=5)
dt_default = DecisionTreeClassifier()
dt_depth3 = DecisionTreeClassifier(max_depth=3)

models = [knn3, knn5, dt_default, dt_depth3]

for model in models:
    model.fit(X_train, y_train)
```

Evaluation Before Noise

```python
knn3_acc = accuracy_score(y_test, knn3.predict(X_test))
knn5_acc = accuracy_score(y_test, knn5.predict(X_test))
dt_def_acc = accuracy_score(y_test, dt_default.predict(X_test))
dt_d3_acc = accuracy_score(y_test, dt_depth3.predict(X_test))
```

Adding Noise to the Test Data

```python
noise = np.random.normal(1, 2.0, X_test.shape)
X_test_noisy = X_test + noise
```

Evaluation After Adding Noise

```python
knn3_acc_n = accuracy_score(y_test, knn3.predict(X_test_noisy))
knn5_acc_n = accuracy_score(y_test, knn5.predict(X_test_noisy))
dt_def_acc_n = accuracy_score(y_test, dt_default.predict(X_test_noisy))
dt_d3_acc_n = accuracy_score(y_test, dt_depth3.predict(X_test_noisy))
```

```python
print("Before Noise:")
print("KNN(K=3):", round(knn3_acc, 2))
print("KNN(K=5):", round(knn5_acc, 2))
print("Decision tree(default):", round(dt_def_acc, 2))
print("Decision tree(depth=3):", round(dt_d3_acc, 2))

print("\nAfter Noise:")
print("KNN(K=3):", round(knn3_acc_n, 2))
print("KNN(K=5):", round(knn5_acc_n, 2))
print("Decision tree(default):", round(dt_def_acc_n, 2))
print("Decision tree(depth=3):", round(dt_d3_acc_n, 2))
```
